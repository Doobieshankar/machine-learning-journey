# pivot_table()

```python
# read and show the dataset
df_sales = pd.read_excel('supermarket_sales.xlsx')
df_sales
```

```python
# make a pivot table and add an aggregate function
#df_sales.pivot_table(index='Gender',aggfunc='sum').round(1)#error because we cant sum date and time but we can get the mean
df_sales.pivot_table(index='Gender',values=['Invoice ID', 'Branch', 'City', 'Customer type',
       'Product line', 'Unit price', 'Quantity', 'Tax 5%', 'Total', 'Payment', 'cogs', 'gross margin percentage', 'gross income',
       'Rating'],aggfunc='min').round(1)
```

```python
df_sales.columns
```

```python
df_sales.describe()
```

```python
# make a pivot table and add an aggregate function
df_sales.pivot_table(index='Gender',
                     values=['Unit price','Quantity','Tax 5%','Total','cogs','gross margin percentage','gross income','Rating'],
                     aggfunc='sum').round(1)
```

```python
# make a pivot table and add an aggregate function
df_sales.pivot_table(index=['Gender','Branch'],
                     values=['Unit price','Quantity','Tax 5%','Total','cogs','gross margin percentage','gross income','Rating','Date'],
                     aggfunc='mean').round(1)
```

```python
# make a pivot table, add an aggregate function and select some columns 
df_sales.pivot_table(index='Gender',
                     values=['Quantity', 'Total'],
                     aggfunc='sum').round(1)
```

```python
# make a pivot tables that says how much male and female spend in each category
df_sales.pivot_table(index='Gender',
                     columns='Product line',
                     values='Total',
                     aggfunc='sum').round(1)
```

```python
# make a pivot tables that says how much male and female spend in each category
df_sales.pivot_table(index=['Gender','Branch'],
                     columns='Product line',
                     values='Total',
                     aggfunc='sum').round(1)
```
